```python
import pandas as pd
import numpy as np
```

#### Задание 1. Загрузка данных
Изучить представленный набор данных на основе описания его столбцов, загрузить его и оставить 8 столбцов для дальнейшего изучения: surgery?, Age, rectal temperature, pulse, respiratory rate, temperature of extremities, pain, outcome.

```python
"""
Тут все просто, обратился, загрузил, дал имена столбцам, заполнил ? на NaN
По описанию нужны 8 столбцов: 0,1,3,4,5,6,10,22
"""

csv = 'https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/statistics_basics/horse_data.csv'
names = ['surgery?', 
         'Age', 
         'rectal temperature', 
         'pulse', 
         'respiratory rate', 
         'temperature of extremities', 
         'pain', 
         'outcome']
horses = pd.read_csv(csv, header=0, usecols=[0,1,3,4,5,6,10,22], names=names, na_values='?')
horses
```

#### Задание 2. Первичное изучение данных
Проанализировать значения по столбцам, рассчитать базовые статистики, найти выбросы.

```python
"""
Не стал отдельно считать многие базовые стастики, а просто сделал describe всего датафрейма
Многие статистики уже посчитались
"""

horses.describe()
```

```python
"""
Оставшиеся посчитал в цикле по каждому столбцу
Использовал список с названием столбцов (был в самом начале) и итеррировался по нему
При этом итерративно по каждому столбцу проверял на выбросы и в remove_outliers писал датафрейм без выбросов
Если встречалось NaN - за выброс не считал, потому что пропуск значения не может быть выбросом
"""

remove_outliers = horses

for col in names:
    print("Медиана", col, "->", horses[col].median())
    print("Размах", col, "->", horses[col].max()-horses[col].min())
    print("Мода", col, "->", horses[col].mode()[0])
    print("Дисперсия", col, "->", horses[col].var())
    print("Межквартильный размах", col, "->", horses[col].quantile(0.75)-horses[col].quantile(0.25))
    print("--------------------------------------------")

    q1 = horses[col].quantile(0.25)
    q3 = horses[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    remove_outliers = remove_outliers[(remove_outliers[col].between(lower_bound, upper_bound, inclusive=True)) | pd.isna(remove_outliers[col])]
```

```python
"""
Затем удалил из исходного датафрейма полученный датафрейм без выбросов в строках
Остался датафрейм с выбросами в строках - т.е. нашел все выбросы
"""

pd.concat([horses, remove_outliers]).drop_duplicates(keep=False)
```

#### Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
"""
Через isna и sum посчитал количество пропусков в каждом столбце
"""

horses.isna().sum()
```

```python
# Перешли к заполнению
# 'temperature of extremities' может быть разной в зависимости была ли операция, каков исход и какой возраст - по этим параметрам сгруппировал и подставил медиану.
# больше ни к чему не смог привязаться, если добавлять боль - оставались пропуски
horses['temperature of extremities'].fillna(horses.groupby(['surgery?','outcome', 'Age'])['temperature of extremities'].transform('median'), inplace=True)
# rectal temperature по условию зависит от temperature of extremities и на ее основании сгруппировал и подставил медиану
horses['rectal temperature'].fillna(horses.groupby('temperature of extremities')['rectal temperature'].transform('median'), inplace=True)
# пульс предположил, что может быть разный в зависимости от была или нет операция, возраста, верхней границы температуры, возраста и исхода
# по этим парамтерам сгруппировал и подставил медиану
horses['pulse'].fillna(horses.groupby(['surgery?', 'temperature of extremities', 'Age', 'outcome'])['pulse'].transform('median'), inplace=True)
# Уровень дыхания - тут сложней всего, оно точно может быть разным у тех кто делал операция, а кто нет
# При разных температурных границах - тоже наверно разный уровень дыхания
# По этим параметрам сгруппировал и подставил медиану
# Больше ни к чему не получилось привязаться - оставались пропуски, да и не факт, что дыхание зависит от боли
horses['respiratory rate'].fillna(horses.groupby(['surgery?', 'temperature of extremities'])['respiratory rate'].transform('median'), inplace=True)
# Боль - точно отличается у тех кто умер, и кто нет, и от того, была ли операция или нет. 
# По этим параметрам сгруппировал и подставил медиану
# Влияние температуры - не факт что есть
horses['pain'].fillna(horses.groupby(['surgery?', 'outcome'])['pain'].transform('median'), inplace=True)
# Наличие операции думаю зависит от температуры, возраста и уровня боли
# По этим параметрам сгруппировал и подставил медиану
# На дыхание написано полагаться нельзя
horses['surgery?'].fillna(horses.groupby(['temperature of extremities', 'Age', 'pain'])['surgery?'].transform('median'), inplace=True)
# Исход думаю зависит от температуры, возраста, уровня боли и у тех кто делал операцию и кто нет - тоже может отличаться исход
# По этим параметрам сгруппировал и подставил медиану
horses['outcome'].fillna(horses.groupby(['surgery?', 'temperature of extremities', 'Age', 'pain'])['surgery?'].transform('median'), inplace=True)
horses
```
