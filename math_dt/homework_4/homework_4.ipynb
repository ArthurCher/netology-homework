```python
# Импортируем нужные библиотеки

from sympy import * 
import matplotlib.pyplot as plt
import numpy as np
```

#### Уровень 0:

Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1.

```python
x = symbols('x')

fx = x**5+4*sin(2*x)+cos(3*x+3)
dfx = diff(fx, x)
print("Первая производная : {}".format(dfx))
x0 = 1
y0 = dfx.evalf(subs={x:x0})
print("Значение первой производной при x = 1 - > ", y0)

ddfx = diff(dfx, x)
print("Вторая производная : {}".format(ddfx))
y0 = ddfx.evalf(subs={x:x0})
print("Значение второй производной при x = 1 - > ", y0)
```

#### Уровень 1:

Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
x = symbols('x')
fx = sin(2*x+1)**5

dfx = diff(fx, x)
print("Первая производная : {}".format(dfx))

ddfx = diff(dfx, x)
print("Вторая производная : {}".format(ddfx))

x_list = []
y_dfx = []
y_ddfx = []

for x0 in np.arange(-5,5,0.01):
    y0 = dfx.evalf(subs={x:x0})
    y1 = ddfx.evalf(subs={x:x0})
    x_list.append(x0)
    y_dfx.append(y0)
    y_ddfx.append(y1)

plt.figure()
plt.plot(x_list, y_dfx)
plt.plot(x_list, y_ddfx)
plt.legend(['Функция первой производной', 'Функция второй производной'])
plt.show()
```
