### Задание.

При фиксированном seed=42 поразбирайтесь и поэкспериментируйте с параметрами алгоритма дифференциальной эволюции: strategy, popsize, tol, mutation, recombination и updating.

Постройте графики количества итераций оптимизации функции ackley от значения параметра.

```python
import numpy as np
from scipy.optimize import differential_evolution
import matplotlib.pyplot as plt
```

```python
def ackley(x):
    arg1 = -0.2 * np.sqrt(0.5 * (x[0] ** 2 + x[1] ** 2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e

bounds = [(-10, 10), (-10, 10)]

result = differential_evolution(ackley, bounds, seed=42)
result
```

```python
# Зависимость от параметра strategy

list_strategy = ['best1bin', 
                 'best1exp', 
                 'rand1exp', 
                 'randtobest1exp',
                 'currenttobest1exp', 
                 'best2exp', 
                 'rand2exp', 
                 'randtobest1bin',
                 'currenttobest1bin',
                 'best2bin',
                 'rand2bin',
                 'rand1bin']

list_nit = []

for strategy in list_strategy:
    result = differential_evolution(ackley, bounds, seed=42, strategy=strategy)
    list_nit.append(result.nit)

plt.figure(figsize=(20, 10))    
plt.bar(list_strategy, list_nit)
plt.show()
```

```python
# Зависиомсть от параметра popsize

list_nit = []
list_popsize = []

for size in range(15, 51, 5):
    result = differential_evolution(ackley, bounds, seed=42, popsize=size)
    list_nit.append(result.nit)
    list_popsize.append(size)

plt.figure(figsize=(20, 10))    
plt.plot(list_popsize, list_nit)
plt.show()
```

```python
# Зависимость от параметра tol

list_nit = []
list_tol = []

for tol in np.arange(0.01, 0.1, 0.005):
    result = differential_evolution(ackley, bounds, seed=42, tol=tol)
    list_nit.append(result.nit)
    list_tol.append(tol)

plt.figure(figsize=(20, 10))    
plt.plot(list_tol, list_nit)
plt.show()
```

```python
# Зависимость от параметра mutation

list_nit = []
list_mutation = []

for mutation in np.arange(0.1, 2, 0.01):
    result = differential_evolution(ackley, bounds, seed=42, mutation=mutation)
    list_nit.append(result.nit)
    list_mutation.append(mutation)
    
plt.figure(figsize=(20, 10))    
plt.plot(list_mutation, list_nit)
plt.show()
```

```python
# Зависимость от параметра recombination

list_nit = []
list_recombination = []

for recombination in np.arange(0.1, 1, 0.005):
    result = differential_evolution(ackley, bounds, seed=42, recombination=recombination)
    list_nit.append(result.nit)
    list_recombination.append(recombination)
    
plt.figure(figsize=(20, 10))    
plt.plot(list_recombination, list_nit)
plt.show()
```

```python
result_immediate = differential_evolution(ackley, bounds, seed=42, updating='immediate')
result_deferred = differential_evolution(ackley, bounds, seed=42, updating='deferred')
print("Количество итерация при updating='immediate' -> ", result_immediate.nit)
print("Количество итерация при updating='deferred' -> ", result_deferred.nit)
```

#### Выводы:
Тут в принципе из графиков все видно. strategy='best1bin является оптимальным с точки зрения количества интерации. Чем больше popsize то как правило больше и количество итераций. tol не влияет на количество итераций. mutation - чем увеличивать - тем как правило и количество интераций увеличивается, но есть интересные значения большие значения mutation, при которых количество интераций также небольшое. С увеличением recombination падает количество интераций. При updating='immediate' - количество итераций меньше.
