### Задание 1:
Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
```

```python
# Нормальное распределение

mu_list = range(1,6)
sigma_list = np.arange(0.1, 0.6, 0.1)

for mu, sigma in zip(mu_list, sigma_list):
    norm_rv = sts.norm(loc=mu, scale=sigma)

    x = np.linspace(0,6,100)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf)

plt.show()
```

```python
# Равномерное распределение

a = 1

for b in range(2,8):
    uniform_rv = sts.uniform(a, b-a)

    x = np.linspace(0,10,100)
    pdf = uniform_rv.pdf(x)
    plt.plot(x, pdf)
plt.show()
```

```python
# Экспоненциальное распределение

for lam in np.arange(0.1, 0.6, 0.1):
    exp_rv = sts.expon(lam)
    x = np.linspace(0,2,100)
    pdf = exp_rv.pdf(x)
    plt.plot(x, pdf)

plt.show()
```

```python
# Т распределение Стьюдента

for n in range(10, 520, 100):
    t_rv = sts.t(n)
    x = np.linspace(-5,5,1000)
    pdf = t_rv.pdf(x)
    plt.plot(x, pdf)

plt.show()
```

### Выводы:
1. Нормальное распределение. При изменении мю логично просто вершина графика (среднее значение) смещается в сторону изменения мю - изменяется среднее значение. При изменении сигмы (средне квадратичного отклонения) естественно разброс от среднего увеличивается, что на графиках видно хорошо
2. Равномерное распределение. При увеличении левой границы распределения соответственно уменьшается верхняя границ чтобы площадь оставалось равно 1
3. При изменении параметра лямбда передвигается вершина графика в сторону его увеличения. Соответственно среднее значение случайной величины увеличивается.
4. Т распределение Стьюдента. Увеличение выборки случайных величин практически не изменяет график. Т.е. среднее значение остается примерно таким же и отклонение от него почти не меняется.

Найдите C, математическое ожидание $\xi$ а также вероятность попадания в отрезок $[3.5;7]$, т.е. P(3.5 < $\xi$ < 7).

### Задание 3
Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

```python
import math

t_red = 54
t_yel = 4
t_green = 28

p1 = t_red / (t_red + t_yel + t_green)
p2 = t_yel / (t_red + t_yel + t_green)
p3 = t_green / (t_red + t_yel + t_green)

I = -(p1*math.log2(p1)+p2*math.log2(p2)+p3*math.log2(p3))

print("Количество информации в сообщении о цвете светофора в текущий момент ->", I)
```

```python
# случае 28, 28 и 28 секунд для каждого цвета
# вероятность каждого события равно 1/3, т.к. всего 3 равновероятных события
# события равно вероятны и применяем формулу Хартли в данном случае

N = 3

I = math.log2(N)
print("Количество информации в сообщении о цвете светофора в текущий момент при этих условиях ->", I)
```
