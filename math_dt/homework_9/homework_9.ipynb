```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.

```python
# rv = sts.norm(1, 0.1) # Пробовал нормальное распределение - на выходе также получается распределение выборочных средних близкое к нормальному
# rv = sts.uniform(1, 4)
rv = sts.expon(0.4) # Оставил для экспоненциального распределения
sample = rv.rvs(size=1000)
```

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
# Параметры запуска: размеры выборок и количество повторов генерации
n1 = 2

samples_count = 1000
# Массивы для сохранения средних по выборкам
for _n in range(n1, 70, 10): # Итерировался по количеству выборок от 2 до 62 с шагом 10
    a1 = np.array([])
    for number in range(samples_count):
        # На каждом цикле генерируем выборки размеров соответствующему размеру итерируемому значению цикла
        sample1 = rv.rvs(size=_n)

        # Считаем среднее по каждой выборке и добавляем в массив средних
        a1 = np.append(a1, sample1.mean())
    
    # Строим графики
    # Подписи для графиков
    label_1 = 'sample of ' + str(_n) + ' histogram'
    plt.hist(a1, density=True, label=label_1)

    plt.legend()
    plt.ylabel('samples means')
    plt.xlabel('$x$')
```

###  Выводы:

Центральная предельная теорема подтвердается. И для нормального распределения случайной величины, и для равномерного, и для экспоненциального - распределение выборочного среднего стремиться к нормальному. При этом чем больше размер выборки, тем меньше дисперсия и тем более явным становится нормальное распределение.
