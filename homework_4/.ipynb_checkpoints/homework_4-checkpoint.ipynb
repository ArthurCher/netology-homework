# Домашнее задание к лекции "Функции"

Данные хранятся в файлах, чтобы при желании данные в них можно было изменить, а программа работала также.
Файлы:
* documents.json - информация о документах сотрудников
* directories.json - информация, о том на каких полках хранятся документы

```python
import json


def get_data_documents():
    with open('documents.json') as data_file:
        documents = json.load(data_file)
    
    return documents
            

def get_data_directories():    
    with open('directories.json') as data_file:
        directories = json.load(data_file)
        
    return directories


def find_user(num_doc):
    documets = get_data_documents()
    name = ''
    
    for document in documets:
        if document['number'] == num_doc:
            name = document['name']

    return name
            

def find_directory(num_doc):
    directories = get_data_directories()
    shelf = ''
    
    for directory in directories:
        if num_doc in directories[directory]:
            shelf = directory
    
    return shelf


def all_info():
    documets = get_data_documents()
    directories = get_data_directories()
    
    for document in documets:
        print("№: ", document['number'], \
              ", тип: ", document['type'], \
              ", владелец: ", document['name'], \
              ", полка хранения: ", find_directory(document['number']),  sep='')
    

def add_directory(num_directory):
    directories = get_data_directories()
    
    if num_directory not in list(directories.keys()):
        directories[num_directory] = []
        
        with open('directories.json', 'w') as data_file:
            json.dump(directories, data_file)
    
    else:
        print("Такая полка уже существует.", end=' ')
        
    print("Текущий перечень полок:", sep='' , end=' ')
    
    step = 1
    
    for shelf in list(directories.keys()):
        if step != len(list(directories.keys())):
            print(shelf, end=', ')
            
        else:
            print(shelf, end='.\n\n')
        
        step += 1

        
def delete_directory(num_directory):
    directories = get_data_directories()
    
    if num_directory in list(directories.keys()):
        if len(directories[num_directory]) == 0:
            del directories[num_directory]
            
            with open('directories.json', 'w') as data_file:
                json.dump(directories, data_file)
            
            print("Полка удалена.", end=' ')
        
        else:
            print("На полке есть документа, удалите их перед удалением полки.", end=' ')
    
    else:
        print("Такой полки не существует.", end=' ')
    
    print("Текущий перечень полок:", sep='' , end=' ')
    
    step = 1
    
    for shelf in list(directories.keys()):
        if step != len(list(directories.keys())):
            print(shelf, end=', ')
            
        else:
            print(shelf, end='.\n\n')
        
        step += 1
                             

def add_document(num_doc, type_doc, name_doc, shelf_doc):
    documents = get_data_documents()
    directories = get_data_directories()
    
    num_docs = []
    
    for document in documents:
        num_docs.append(document['number'])
    
    if num_doc in num_docs:
        print("Такой документ уже есть в базе. Текущий список документов:\n")
        
    
    elif shelf_doc not in list(directories.keys()):
        print("Такой полки не существует. Добавьте полку командой as.\nТекущий список документов:\n")
        
    else:
        new_document = {}
        
        new_document['type'] = type_doc
        new_document['number'] = num_doc
        new_document['name'] = name_doc
        
        documents.append(new_document)
        
        with open('documents.json', 'w') as data_file:
            json.dump(documents, data_file)
        
        directories[shelf_doc].append(num_doc)
        
        with open('directories.json', 'w') as data_file:
            json.dump(directories, data_file)
            
        print("Документ добавлен.\nТекущий список документов:\n")
        
    print('\n')
    all_info()
    

def delete_document(num_documents):
    documents = get_data_documents()
    directories = get_data_directories()
    
    num_docs = []
    
    for document in documents:
        num_docs.append(document['number'])
    
    if num_documents in num_docs:
        for document in documents:
            if document['number'] == num_documents:
                documents.remove(document)

        with open('documents.json', 'w') as data_file:
            json.dump(documents, data_file)

        for directory in directories:
            if num_documents in directories[directory]:
                directories[directory].remove(num_documents)

        with open('directories.json', 'w') as data_file:
                json.dump(directories, data_file)
        
        print("\nДокумент удален.\nТекущий список документов:\n")
    
    else:
        print("\nДокумент не найден в базе.\nТекущий список документов:\n")
    
    all_info()
    

def move_document(num_documents, num_directory):
    documents = get_data_documents()
    directories = get_data_directories()
    
    num_docs = []
    
    for document in documents:
        num_docs.append(document['number'])
    
    if num_doc not in num_docs:
        print("\nТакого документа нет в базе. Текущий список документов:\n")
        
        all_info()        
    
    elif num_directory not in list(directories.keys()):
        print("\nТакой полки не существует. Добавьте полку командой as.")
        
        print("Текущий перечень полок:", sep='' , end=' ')
    
        step = 1
    
        for shelf in list(directories.keys()):
            if step != len(list(directories.keys())):
                print(shelf, end=', ')
            
            else:
                print(shelf, end='.\n\n')
        
        step += 1
    
    else:
        for directory in directories:
            if num_documents in directories[directory]:
                directories[directory].remove(num_documents)
        
        for directory in directories:
            if num_directory == directory:
                directories[directory].append(num_documents)
        
        with open('directories.json', 'w') as data_file:
            json.dump(directories, data_file)
        
        print("\nДокумент перемещен.\nТекущий список документов:\n")
        
        all_info()



print("Утилита работа с документами")

while True:
    to_do = input("\nВведите желаемое действие:\n \
    Введите p (англ.) - для поиска сотрудника по номеру документа\n \
    Введите s - для поиска полки, где лежит документ\n \
    Введите l - для вывода все информации по документам\n \
    Введите ads - для добавления новой полки\n \
    Введите ds - для удаление пустой полки\n \
    Введите ad - для добавления нового документа\n \
    Введите d - для удаления документа\n \
    Введите m - для перемещения документа на другую полку\n \
    Введите q - для выхода их программы\n\n")
    
    if to_do == 'q':
        break
        
    elif to_do == 'p':
        num_doc = input("Введите номер документа - > ")
        name_users = find_user(num_doc)
        
        print("Резултат: ")
        
        if name_users != '':
            print("Владелец документа: ", name_users, "\n")
        
        else:
            print("Документ не найден в базе\n")            
    
    elif to_do == 's':
        num_doc = input("Введите номер документа - > ")
        doc_directory = find_directory(num_doc)
        
        print("Резултат: ")
        
        if doc_directory != '':
            print("Документ хранится на полке: ", doc_directory, "\n")
        
        else:
            print("Документ не найден в базе\n")
        
    
    elif to_do == 'l':
        all_info()
        
    
    elif to_do == 'ads':
        new_shelf = input("Введите номер новой полки - > ")
        add_directory(new_shelf)
        
    
    elif to_do == 'ds':
        del_shelf = input("Введите номер полки, которую хотите удалить - > ")
        delete_directory(del_shelf)
    
    
    elif to_do == 'ad':
        num_doc = input("Введите номер документа - > ")
        type_doc = input("Введите тип документа - > ")
        name_doc = input("Введите владельца документа - > ")
        shelf_doc = input("Введите полку для хранения - > ")
        
        add_document(num_doc, type_doc, name_doc, shelf_doc)
        
    
    elif to_do == 'd':
        num_doc = input("Введите номер документа, который нужно удалить - > ")
        
        delete_document(num_doc)
        
    
    elif to_do == 'm':
        num_doc = input("Введите номер документа, который нужно переместить - > ")
        new_shelf = input("Введите номер полки, куда нужно переместить документ - > ")
        
        move_document(num_doc, new_shelf)
        
    
    else:
        print("Введена не верная команда")
        
        
    repet = input("Для продолжения работы введите любой символ или нажмите ввод, для прекращения введите q - > ")
    
    if repet == 'q':
        break
```
