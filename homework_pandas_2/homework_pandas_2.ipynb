#### Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
"""
Считал в ratings весь файл с оценками фильмов.
Оставил только стобцы movieId и rating
Затем в одно действие 
1) фильтранул с оценками 5;
2) сгруппировал по movieId и для каждой группы посчитал количество оценок
3) отсортировал по убыванию, чтобы фильм с наибольшим количеством оценок был первым
4) вывел только первую строку и записал все это в новый дата фрейм
Затем приджойнил к этому новому датафрейму данные из файла с названиями фильмов
Так вывелось название фильма с наибольшим количеством оценок 5

Пометка в папке ml-25m файлы заменил файлами из ml-latest-small.zip - git отказался такие больший файлы заливать
"""

ratings = pd.read_csv('ml-25m/ratings.csv')
movies_rating = ratings[['movieId','rating']]
result_film = movies_rating[movies_rating['rating'] == 5].groupby('movieId').count().sort_values(by='rating', ascending=False).head(1)

movies = pd.read_csv('ml-25m/movies.csv')
result_film.merge(movies, on='movieId', how='left')
```

#### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
"""
Все считал в countries
Потом начал фильтровать по очереди и записывать результат в filtered_countries:
1) Отфильтровал по категории 4, 12 и 21
2) Отфильтровал по названию стран Latvia, Lithuania, Estonia
3) Отфильтровал по году с 2005 по 2010 года
4) Отфильтровал по quantity > 0
Далее оставил только стоблцы country, quantity, сгруппировал по стране и для каждой группы посчитал сумму quantity
"""

countries = pd.read_csv('power.csv')
filtered_countries = countries[(countries['category'] == 4) | (countries['category'] == 12) | (countries['category'] == 21)]
filtered_countries = filtered_countries[(filtered_countries['country'] == 'Latvia') | (filtered_countries['country'] == 'Lithuania') | (filtered_countries['country'] == 'Estonia')]
filtered_countries = filtered_countries[(filtered_countries['year'] >= 2005) & (filtered_countries['year'] <= 2010)]
filtered_countries = filtered_countries[filtered_countries['quantity'] > 0]
filtered_countries[['country', 'quantity']].groupby('country').sum('quantity')
```
