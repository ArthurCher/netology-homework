```python
games = pd.read_csv('https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/stat_case_study/vgsales.csv')
sport_games = games.loc[games['Genre'] == 'Sports']
sport_games['Critic_Score'].min(), sport_games['Critic_Score'].max()
```

#### Как критики относятся к спортивным играм? 
#### Гипотезы:
H0: средняя удовлетворенность критиков спортивными играми <= 80% \
H1: средняя удовлетворенность критиков спортивными играми более 80%

За alpha будем принимать 0.05

```python
alpha = 0.05
result = stats.ttest_1samp(sport_games['Critic_Score'], 80, alternative='greater')

if result.pvalue < alpha: 
    print('Отвергаем нулевую гипотезу, средняя оценка больше 80 (спортивные игры нравятся)')
else:
    print('Не отвергаем нулевую нулевую гипотезу, спортивные игры не нравятся критикам')
```

#### Критикам нравятся больше игры на PC или на PS4? 

H0: Критикам одинаково нравятся игры на PC и на PS4 \
H1: Критикам не одинаково нравятся игры на PC и на PS4

За alpha будем принимать 0.05

```python
pc_games = games.loc[games['Platform'] == 'PC']
pc_games = pc_games.loc[:,['Name', 'Platform', 'Critic_Score']]
pc_games.dropna(inplace=True)

ps4_games = games.loc[games['Platform'] == 'PS4']
ps4_games = ps4_games.loc[:,['Name', 'Platform', 'Critic_Score']]
ps4_games.dropna(inplace=True)

t , p  = stats.ttest_ind(pc_games['Critic_Score'], ps4_games['Critic_Score'])
t , p

alpha = 0.05
print('P-value ->', p)
print('T-score ->', t)


if p > alpha:
    print('Одинаковое распределение (не отвергаем H0)')
else:
    print('Разное распределение (отклоняем H0)')
```

```python
# Визуализируем разницу оценок
# Сделаем 1000 выборок из этих множеств, посчитаем среднее и построим их распределение

plt.figure(figsize=(15, 10))

boot_1d_pc = []
boot_1d_ps4 = []
for i in range(1000):
    boot_mean_pc = pc_games['Critic_Score'].sample(frac = 1, replace = True).mean()  
    boot_1d_pc.append(boot_mean_pc)
    boot_mean_ps4 = ps4_games['Critic_Score'].sample(frac = 1, replace = True).mean() 
    boot_1d_ps4.append(boot_mean_ps4)

boot_1d = {'mean_pc_score': boot_1d_pc, 'mean_ps4_score': boot_1d_ps4}

boot_1d = pd.DataFrame(boot_1d)
   
boot_1d.plot(kind='density')
plt.show()
```

#### Ответ на вопрос:

Критикам больше нравятся игры на PC

#### Критикам больше нравятся стрелялки или стратегии?

H0: Критикам одинаково нравятся стрелялки и стратегии \
H1: Критикам не одинаково нравятся стрелялки и стратегии

За alpha будем принимать 0.05

```python
shooter_games = games.loc[games['Genre'] == 'Shooter']
shooter_games = shooter_games.loc[:,['Name', 'Genre', 'Critic_Score']]
shooter_games.dropna(inplace=True)

strategy_games = games.loc[games['Genre'] == 'Strategy']
strategy_games = strategy_games.loc[:,['Name', 'Genre', 'Critic_Score']]
strategy_games.dropna(inplace=True)

t , p  = stats.ttest_ind(shooter_games['Critic_Score'], strategy_games['Critic_Score'])
t , p

alpha = 0.05
print('P-value ->', p)
print('T-score ->', t)


if p > alpha:
    print('Одинаковое распределение (не отвергаем H0)')
else:
    print('Разное распределение (отклоняем H0)')
```

```python
# Визуализируем разницу оценок
# Сделаем 1000 выборок из этих множеств, посчитаем среднее и построим их распределение

plt.figure(figsize=(15, 10))

boot_1d_shooter = []
boot_1d_strategy = []
for i in range(1000):
    boot_mean_shooter = shooter_games['Critic_Score'].sample(frac = 1, replace = True).mean()  
    boot_1d_shooter.append(boot_mean_shooter)
    boot_mean_strategy = strategy_games['Critic_Score'].sample(frac = 1, replace = True).mean() 
    boot_1d_strategy.append(boot_mean_strategy)

boot_1d = {'mean_shooter_score': boot_1d_shooter, 'mean_strategy_score': boot_1d_strategy}

boot_1d = pd.DataFrame(boot_1d)
   
boot_1d.plot(kind='density')
plt.show()
```

#### Ответ на вопрос:

Критикам больше нравятся стратегии

```python
# Подгрузим все необходимые пакеты и датафрейм

import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from gensim import corpora
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix

df = pd.read_csv('https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/stat_case_study/spam.csv')
df.head(20)
```

```python
# Создадим необходимые функции для преобразования данных в датафреме

# Функция для преобразования текста письма (Убирает стоп-слова и приводит к нормальной форме)
def get_clear_message(message):
    list_words = message.split()
    stopwords_set = set(stopwords.words('english'))
    wordnet_lemmatizer = WordNetLemmatizer()
    list_wors_sw = []

    for word in list_words:
        if word in stopwords_set:
            list_words.remove(word)
            
        else:
            list_wors_sw.append(wordnet_lemmatizer.lemmatize(word))
    
    message_sw = ' '.join(list_wors_sw)
    
    return message_sw

# Создаем модель TF-IDF и получаем датафрейм с раскладкой слов сообщений на векторы
def get_tfidf(df):
    tfidf = TfidfVectorizer()
    tfidf_matrix = tfidf.fit_transform(df['Message'])
    names = tfidf.get_feature_names()
    tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
    
    return tfidf_matrix

# Переименовывает названий категория из текста в число и обратно
def rename_cat(category):
    if category == 'spam':
        return 0
    elif category == 'ham':
        return 1
    elif category == 0:
        return 'spam'
    elif category == 1:
        return 'ham'
    else:
        return False
```

```python
# Сохраним исходные датафрейм отдельно для последующего вывода сообщений с ошибочно предсказанной категорией
df_message = df

# Преобразуем текст сообщения
df['Message'] = df['Message'].apply(lambda text: text.lower()) # Приводим в нижний регистр
df['Message'] = df['Message'].apply(lambda text: re.sub('[\W_]+',' ', text)) # Убираем спец символы
df['Message'] = df['Message'].apply(lambda text: get_clear_message(text)) # Убираем стоп-слова и приводим к нормальной форме
df = pd.concat([df, get_tfidf(df)], axis=1) # Добавляем к исходному датафрейму датафрейм с раскладкой слов сообщений на векторы
df['Category'] = df['Category'].apply(lambda cat: rename_cat(cat)) # Переименуем категории: спам - 0; неспам = 1
df.drop('Message', axis=1, inplace=True) # Уберем столбец с текстом сообщения для последующей работы
df
```
