```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

#### Обязательная часть

Вам необходимо провести базовый EDA выбранного набора данных. 

Требования к анализу:

* построить не менее 4 визуализаций различных видов;
* каждая визуализация должным образом оформлена и читается даже в отрыве от контекста;
* по каждой визуализации необходимо написать вывод (какую гипотезу на ее основе можно выдвинуть?).

```python
# Взял данные об уровне счастья в разных странах.
names = ['Место', 'Страна', 'балл', 'ВВП на душу населения', 'Социальная поддержка', 'Продолжительность жизни', 'Свобода выбора', 'Великодушие', 'Восприятие коррупции']
happiness_rate = pd.read_csv('2019.csv', names=names, header=0)
happiness_rate
```

```python
# Построил зависимость ВВП на душу населения, 'Социальная поддержка, Продолжительность жизни от места в рейтинге

plt.figure(figsize=(15, 10))
names_ = ['ВВП на душу населения', 'Социальная поддержка', 'Продолжительность жизни']

for item in names_:
    plt.plot(happiness_rate.index, happiness_rate[item])
    plt.title('Динамика показателей')
    plt.xlabel('Место в рейтинге')
    plt.ylabel('Значение показателей') 
    plt.legend(names_)
```

Из графика можно сделать следующий вывод: \
Чем больше место в рейтинге тем ниже показатели страны: ВВП на душу населения, Социальная поддержка, Продолжительность жизни

```python
# Отобразил распределение значений Свободы выбора относительно мест в рейтинге

happiness_rate.plot(kind='scatter', 
                    x='Место', 
                    y='Свобода выбора',
                    title='Распределение свободы выбора по местам',
                    figsize=(15, 10))
```

Выводы: \
Видно, что при перемещении по рейтингу сверху вниз, значения свободы выборы приобретают сильный разброс. Т.е. низкое место в рейтинге еще не значит, что в стране низкая свобода выбора. Страны с низкими уровнем счастья могут практически с равной вероятность могут иметь как высокие значения свободы выбора, так и низкие. Но если страна занимает высокое место в рейтинге - однозначно она имеет высокий показатель сводободы выбора

```python
# построил зависимость продолжительности жизни от социальной поддержки 

soc_life = happiness_rate[['Социальная поддержка', 'Продолжительность жизни']].sort_values('Социальная поддержка')

soc_life.plot(kind='area', 
                    x='Социальная поддержка', 
                    y='Продолжительность жизни',
                    title='Зависимость социальной поддержки от продолжительности жизни',
                    figsize=(15, 10))
```

Выводы: \
График показывает, что чем выше социальная поддержка - тем выше продолжительность жизни
