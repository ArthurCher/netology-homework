#### Задание 1**
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:\
* оценка 2 и меньше - низкий рейтинг
* оценка 4 и меньше - средний рейтинг
* оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class\
\**Домашнее задание к лекции.

```python
import pandas as pd
```

```python
"""
Сделал все как в условии, только приджойнил датафрейм с названием фильмов, чтобы было вывелось название фильма
Я подумал так наглядней. Ну и вывел 50 первых строк, чтобы попались фильмы всех рейтингов
"""

def classification_film(rating):
    if rating <= 2:
        return 'низкий рейтинг'
    elif rating <= 4:
        return 'средний рейтинг'
    else:
        return 'высокий рейтинг'

film_data = pd.read_csv('ml-latest-small/ratings.csv')
movies = pd.read_csv('ml-latest-small/movies.csv')
film_data['classification'] = film_data['rating'].apply(classification_film)
joined_data = film_data.merge(movies, on='movieId', how='left')
joined_data.head(50)
```

#### Задание 2
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

```python
"""
Сделал как в условии. 
Только подумал следующее: если использовать в лоб, то запросы, где регионы склоняются - не найдутся.
Например 'москве' при поиске в лоб не найдется и будет undefined.
Подумал что так неправильно. Решил модернизировать поиск.
Вы в одном вебинаре рассказывали про расстояние Левенштейна, решил, что можно попробовать ее использовать.
По идее расстояние Левенштейна должно быть либо 0, либо 1 
Потому что 'москве' от 'москва' отличается на один символ - тоже самое по другим регионам
Соответственно, если в запросе есть слов,
у которого расстояние Левенштейна с искомым городом ровно 0 или 1 - нужный запрос
Вот в циклах перебрал все и сделал простую проверку на это расстояние
Для наглядности вывел для региона 'петербург' - работает
"""


import Levenshtein as lvst

geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
           }

def geo_classificator(keyword):
    list_keyword = keyword.split(' ')
    for region in geo_data:
        for city in geo_data[region]:
            for item in list_keyword:
                if lvst.distance(item, city) <= 1:
                    return city
    
    return 'undefined'

keywords = pd.read_csv('ml-latest-small/keywords.csv')
keywords['region'] = keywords['keyword'].apply(geo_classificator)
filtered_keywords = keywords[keywords['region'] == 'петербург']
filtered_keywords
```

#### Задание 3 (бонусное)

Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл movies.csv и ratings.csv из базы https://grouplens.org/datasets/movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

    1. В переменную years запишите список из всех годов с 1950 по 2010.

    2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
        1. для каждой строки пройдите по всем годам списка years
        2. если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
        3. если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год

    3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

    4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга


```python
"""
Сделал как в условии:
Сделал функцию production_year, передал в нее название фильма и статикой список годов
Сделал проверку и вернул год из списка, если тру, если ничего не нашлось - 1900
Применил функцию на столбце title и записал в новый столбец year этого датафрейма
Сгруппировал по year с подсчетом среднего значения рейтинга фильмов за этот год
Сбросил индекс, отсортировал по убыванию рейтинга и вывел 60 первых строк - потому что всего получилось 60 лет
(так вывелось все)
"""

def production_year(title, years = list(range(1950,2011))):
    for year in years:
        if str(year) in title:
            return int(year)
        else:
            pass
    
    return 1900
    
film_data = pd.read_csv('ml-latest-small/ratings.csv')
movies = pd.read_csv('ml-latest-small/movies.csv')
joined_data = film_data.merge(movies, on='movieId', how='left')
joined_data['year'] = joined_data['title'].apply(production_year)

joined_data.groupby('year').mean()['rating'].reset_index().sort_values('rating', ascending=False).head(60)
```
